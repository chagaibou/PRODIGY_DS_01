```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset of Adult Census Income
df = pd.read_csv('./datasets/adult.csv')
df.head()
```

```python
#Historgram for the distribution of Age
plt.figure()
sns.histplot(df['age'],kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.show()
```

Distribution of Sex

```python
#barchart for the distribution of Sex
plt.figure()
sns.countplot(data=df,x=df['sex'], palette="Set2",hue=df['sex'])
plt.title("Distribution of Sex")
plt.xlabel("Sex")
plt.show()
```
